```python
from pathlib import Path

ds_path = Path(r"/mnt/walkure_public/users/tomererez/mimic_cxr_jpg_images/mimic_cxr_jpg_images_from_google_cloud/mimic-cxr-jpg-2.1.0.physionet.org/files")

# Fast + memory-efficient: iterate paths without building a huge list
jpg_count = sum(
    1
    for p in ds_path.rglob("*")
    if p.is_file() and p.suffix.lower() in (".jpg", ".jpeg")
)

print(jpg_count)
```
