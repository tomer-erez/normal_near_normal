```python
import pandas as pd
df=pd.read_csv(r"C:\Users\tomer.erez\Desktop\code_projects\normal_near_normal\cxr_data\all_txt_data_and_labels.csv")
```

```python
df.columns()
```

```python
import matplotlib.pyplot as plt

chexpert_cols = [col for col in df.columns if col.startswith("chexpert_")]

summary = []

for col in chexpert_cols:
    # Raw counts including NaN
    counts = df[col].value_counts(dropna=False)

    total = len(df)

    num_pos = counts.get(1, 0)
    num_neg = counts.get(0, 0)
    num_uncertain = counts.get(-1, 0)
    num_na = df[col].isna().sum()

    # Normalized
    norm_pos = num_pos / total
    norm_neg = num_neg / total
    norm_uncertain = num_uncertain / total
    norm_na = num_na / total
    col=col.split("chexpert_")[1]  # Extract pathology name for better display
    print(f"\n{col}:")
    print(f"Total samples: {total}")
    print(f"1   -> count={num_pos:6d}, ratio={norm_pos:.4f}")
    print(f"0   -> count={num_neg:6d}, ratio={norm_neg:.4f}")
    print(f"-1  -> count={num_uncertain:6d}, ratio={norm_uncertain:.4f}")
    print(f"NaN -> count={num_na:6d}, ratio={norm_na:.4f}")

    summary.append({
        "column": col,
        "1": num_pos,
        "0": num_neg,
        "-1": num_uncertain,
        "NaN": num_na,
        "1_ratio": norm_pos,
        "0_ratio": norm_neg,
        "-1_ratio": norm_uncertain,
        "NaN_ratio": norm_na
    })

    # Plot
    plt.figure(figsize=(6, 4))

    labels = ["1", "0", "-1", "NaN"]
    values = [num_pos, num_neg, num_uncertain, num_na]

    plt.bar(labels, values)
    plt.title(col)
    plt.ylabel("Count")

    for i, v in enumerate(values):
        plt.text(i, v, str(v), ha='center', va='bottom')

    plt.show()

# Optional summary dataframe
summary_df = pd.DataFrame(summary)

print("\nSummary table:")
print(summary_df)
```
